# 일상용어 조회

173602개 일상용어

```python
common_term = []

id = 'leegun9371'
# https://www.law.go.kr/DRF/lawSearch.do?OC=leegun9371&target=prec&type=XML&query=%ED%8F%AD%EB%A0%A5&display=100&page=2

limit_page = 1800
for i in range(1, limit_page+1):

  print(i)

  # query 부분에 url 한글 적기
  url = f'https://www.law.go.kr/DRF/lawSearch.do?OC={id}&target=dlytrm&type=XML&display=100&page={i}'

  result = urlopen(url)
  house = BeautifulSoup(result, 'lxml-xml')
  te = house.find_all('일상용어')

  for i in range(len(te)):
    term_name = te[i].일상용어명.string.strip()
    term_source = te[i].출처.string.strip()
    term_link = te[i].용어간관계링크.string.strip()

    data = [term_name, term_link]
    common_term.append(data)
```

```python
df_common_term = pd.DataFrame(common_term, columns=['term_name', 'term_link'])
```

```python
from urllib.error import HTTPError
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

voice_result= ['방한귀덮개','검정장비','구제사무','예비검사비']

desired_keywords = voice_result
filtered_indexes = []

# DataFrame을 순회하면서 조건을 확인합니다.
for index, row in df_common_term.iterrows():
    term_name = row['term_name']

    # 'term_name'에 desired_keywords 리스트에 존재하는 문자열이 있는지 확인
    if any(keyword in term_name for keyword in desired_keywords):
        # 존재한다면 해당 행의 index를 저장
        filtered_indexes.append(index)

# 출력
print("\n조건을 만족하는 index:")
print(filtered_indexes)
```
