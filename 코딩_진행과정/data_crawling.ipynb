# Open API

```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
datas = []

id = 'leegun9371'
# https://www.law.go.kr/DRF/lawSearch.do?OC=leegun9371&target=prec&type=XML&query=%ED%8F%AD%EB%A0%A5&display=100&page=2

limit_page = 23
for i in range(1, limit_page+1):

  print(i)

  # query 부분에 url 한글 적기
  url = f'https://www.law.go.kr/DRF/lawSearch.do?OC={id}&target=prec&type=XML&query=%EC%82%AC%EA%B8%B0&display=100&page={i}'

  result = urlopen(url)
  house = BeautifulSoup(result, 'lxml-xml')
  te = house.find_all('prec')

  for i in range(len(te)):
    serial_number = te[i].판례일련번호.string.strip()
    case_name = te[i].사건명.string.strip()
    case_number = te[i].사건번호.string.strip()
    judge_date = te[i].선고일자.string.strip()
    court_name = te[i].법원명.string.strip()
    case_type = te[i].사건종류명.string.strip()
    case_type_code = te[i].사건종류코드.string.strip()
    judge_type = te[i].판결유형.string.strip()
    judge = te[i].선고.string.strip()

    data = [serial_number, case_name, case_number, judge_date, court_name, case_type, case_type_code, judge_type, judge]
    datas.append(data)
```

```python
data2 = []
for i, data in enumerate(datas):

  print(i)

  url = 'http://www.law.go.kr/DRF/lawService.do?OC=leegun9371&target=prec&type=XML&ID=' + data[0]
  result = urlopen(url)
  house = BeautifulSoup(result, 'lxml-xml')
  te = house.find('PrecService')

  serial_number = te.판례정보일련번호.string.strip()
  case_name = te.사건명.string.strip()
  case_number = te.사건번호.string.strip()
  judge_date = te.선고일자.string.strip()
  judge = te.선고.string.strip()
  court_name = te.법원명.string.strip()
  court_type_code = te.법원종류코드.string.strip()
  case_type = te.사건종류명.string.strip()
  case_type_code = te.사건종류코드.string.strip()
  #result_type = te.판시유형.string.strip()
  result = te.판시사항.string.strip()
  result_interpret = te.판결요지.string.strip()
  reference = te.참조조문.string.strip()
  reference_past = te.참조판례.string.strip()
  past_detail = te.판례내용.string.strip()

  data = [serial_number, case_name, case_number, judge_date, judge, court_name, court_type_code, case_type, case_type_code,
          result, result_interpret, reference, reference_past, past_detail]
  data2.append(data)
```
